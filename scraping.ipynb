```python
# Importing the libraries that we will need.
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Defining the wikipedia url that we will scrape.
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
# Getting the page content and parsing it with BeautifulSoup.
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
# Finding the table that we want to scrape.
table = soup.find(id="constituents")
print(table.prettify())
```

```python
# Getting the table headers.
table_headers = table.find_all("th")
# Transforming them into a list.
table_headers = [i.text for i in table_headers]
# Removing the \n from the list.
table_headers = [i.replace("\n", "") for i in table_headers]
print(table_headers)
```

```python
# Getting the data in the table.
table_data = table.find_all("td")
# Transforming them into a list.
table_data = [item.text for item in table_data]
# Removing the \n from the list.
table_data = [item.replace("\n", "") for item in table_data]
# Grouping the rows into lists of the number of elements of the headers.
table_data = [table_data[i:i+len(table_headers)] for i in range(0,len(table_data),len(table_headers))]

table_data
```

```python
# Creating a dataframe with the data.
df = pd.DataFrame(table_data, columns=table_headers)
df
```

```python
# Saving the dataframe as a csv file.
df.to_csv("sp500.csv", index=False)
```
